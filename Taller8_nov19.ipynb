Gr√°fico cuantil-cuantil
=======================

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.random.normal(10,2,1000)
y=np.random.normal(-2,5,10)
```

```python
q=np.linspace(0,1,len(y))
xq=np.quantile(x,q)
yq=np.quantile(y,q)
```

```python
plt.scatter(xq,yq)
```

```python
# O m√°s rapidamente a trav√©s de sicpy
from scipy import stats

res = stats.probplot(y, plot=plt)
```
